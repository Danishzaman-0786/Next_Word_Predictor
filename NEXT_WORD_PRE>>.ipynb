```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
Tokenizer=Tokenizer()
```

```python
Tokenizer.fit_on_texts([Data_])
```

```python
(Tokenizer_.word_index)
```

```python
type(Data_)
```

```python
#we need to split the senteses
for i in Data_.split('\n'):
  print(i)
```

```python
#now we find the numbers format
input_data=[]
for i in Data_.split('\n'):
  p_=(Tokenizer.texts_to_sequences([i])[0])
  for i in range(1,len(p_)):
     input_data.append(p_[:i+1])
```

```python
input_data
```

```python
#how to find the maximum lenght of sentense
max_=max([len(i) for i in input_data])
max_
```

```python
#maximum lenght_sentes=57
#we use the zero padding
from tensorflow.keras.preprocessing.sequence import pad_sequences
input_padding=pad_sequences(input_data,maxlen=max_,padding='pre')
```

```python
input_padding
```

```python
X=input_padding[:,:-1]
#we use another method

```

```python
X
```

```python
#as well as using for y
Y=input_padding[:,-1]
```

```python
Y
```

```python
Y.shape
```

```python
#we use the OHT
from tensorflow.keras.utils import to_categorical
Y=to_categorical(Y,num_classes=len(Tokenizer.word_index)+1)
```

```python
Y
```

```python
Y.shape
```

```python
#now we have X,Y then i build the Architecture of LSTM
#we first layer(embedding),2nd(LSTM),(dense_layer)..this main ARchtitecture
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model=Sequential()
```

```python
#Embedding(total_words,total_nos_of _dense_in_output_vector,which_setnse_have_maximum_length_in_the_given_data)
```

```python
model.add(Embedding(len(Tokenizer.word_index)+1,100,input_length=57))
```

```python
#we design a LSTM ...LSTM(NOs_of_nodes)
model.add(LSTM(150,return_sequences=False))
```

```python
#we design dense....dense(totl_nos_of_word)
model.add(Dense(len(Tokenizer.word_index)+1,activation='softmax'))
#which word has highest probability his we assign the 1..example...p[0.2,0.5,0.6]..max(p)=0.6..then..result:-:[0,0,1]_in the same
```

```python
#Now we do compile our model
#model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#Now we seen the summary of our model
model.summary()
```

```python
model.fit(X,Y,epochs=10)
```

```python
#Now we startb the prediction
#let's..Data1='mail'
#we take a tokenizer
#padding
#prediction
```

```python
import time
n_test_data='data science'
for i in range(10):
    tokenizer1=Tokenizer.texts_to_sequences([n_test_data])[0]
    padding_=pad_sequences([tokenizer1],maxlen=57,padding='pre')
    #our model gives the probebility for multiple classes
    #then whic is highest probability then =1
    #now use argmax in numpy for tell us that higher probability which position
    import  numpy as np
    word_position=np.argmax(model.predict(padding_))
    for i,j in Tokenizer.word_index.items():
      if j==word_position:
        n_test_data=n_test_data+' '+i
        print(n_test_data)
        time.sleep(2)
```
